```python
import pandas as pd
import numpy as np

khdata = "/Users/ritchevy/Desktop/khdata.csv"
kh = pd.read_csv(khdata)

kh.head()
```

```python
kh.describe()
```

```python
import matplotlib.pyplot as plt
#kh.hist(bins=50, figsize=(20,15))
kh.plot(x='Name', y=['Global_Sales','Japan_Sales','North_America_Sales','Europe_Sales'], kind='bar')
plt.xlabel('Game name')
plt.ylabel('Sales in Millions')

plt.rcParams["figure.figsize"] = [7,4]
#plt.xticks(kh[['Name']], rotate = 45)
plt.show()
```

```python
kh_region = kh[['Japan_Sales','North_America_Sales', 'Europe_Sales', 'Other_Sales']]
kh_region.sum().sort_values(ascending=False).plot(kind="bar")
plt.xlabel('Sales by Region')
plt.show()
```

```python
#SALES PER REGION
#Sales in Japan
kh.sort_values(by=['Japan_Sales'] , ascending=False).plot(x='Name', y=['Japan_Sales'], kind='bar')
plt.xlabel('Game names')
plt.ylabel('Sales in Millions')
plt.show()
#Sale in North America
kh.sort_values(by=['North_America_Sales'] , ascending=False).plot(x='Name', y=['North_America_Sales'], kind='bar')
plt.xlabel('Game names')
plt.ylabel('Sales in Millions')
plt.show()
#Sale in Europe
kh.sort_values(by=['Europe_Sales'] , ascending=False).plot(x='Name', y=['Europe_Sales'], kind='bar')
plt.xlabel('Game names')
plt.ylabel('Sales in Millions')
plt.show()
```

```python
#Sale in Others
kh.sort_values(by=['Other_Sales'] , ascending=False).plot(x='Name', y=['Other_Sales'], kind='bar')
plt.xlabel('Game names')
plt.ylabel('Sales in Millions')
plt.show()
```
